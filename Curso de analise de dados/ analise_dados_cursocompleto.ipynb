## Atividade Extra:

como fazer isso? Reiniciar o índice ao filtrar dados (reset_index)

Remover ou renomear colunas (.drop(columns=[]) e .rename(columns={}))

Remover valores duplicados e/ou ausentes (.drop_duplicates() e .dropna())

Tratar valores ausentes, substituindo nulos por zero, por exemplo (.fillna(0))

Converter tipos de dados (.astype())

Criar, dividir ou reordenar colunas (para dividir: .split())

Remover espaços em branco (.strip())

Classificar os dados (.sort_values(by='coluna', ascending=False))

#explicações e exemplos:

##1. Reiniciar o índice ao filtrar dados (reset_index)

Após filtrar um DataFrame do Pandas, o índice original é mantido. Para reiniciá-lo numericamente, utilize reset_index():

dados_filtrados = dados[dados['IDADE'] > 30]
dados_filtrados = dados_filtrados.reset_index(drop=True)  # drop=True remove o índice antigo

##2. Remover ou renomear colunas (.drop(columns=[]) e .rename(columns={}))

Remover colunas:
dados = dados.drop(columns=['GENERO', 'IDADE'])

Renomear colunas:
dados = dados.rename(columns={'GENERO': 'Sexo', 'IDADE': 'Idade'})

##3. Remover valores duplicados e/ou ausentes (.drop_duplicates() e .dropna())

Remover linhas duplicadas:
dados = dados.drop_duplicates()


Remover linhas com valores ausentes:
dados = dados.dropna()

##4. Tratar valores ausentes, substituindo nulos por zero, por exemplo (.fillna(0))

dados['IDADE'] = dados['IDADE'].fillna(0)


##5. Converter tipos de dados (.astype())
dados['IDADE'] = dados['IDADE'].astype(int)


##6. Criar, dividir ou reordenar colunas (para dividir: .split())
Criar uma nova coluna a partir de outra, dividindo os valores de uma string usando um delimitador, como espaço em branco.

Supondo que 'NOME_COMPLETO' contenha nome e sobrenome separados por espaço.
dados[['Nome', 'Sobrenome']] = dados['NOME_COMPLETO'].str.split(' ', n=1, expand=True)

Reordenar colunas:
colunas_desejadas = ['Nome', 'Sobrenome', 'IDADE', 'GENERO']
dados = dados[colunas_desejadas]

##7. Remover espaços em branco (.strip())
dados['GENERO'] = dados['GENERO'].str.strip()


##8. Classificar os dados (.sort_values(by='coluna', ascending=False))
dados = dados.sort_values(by='IDADE', ascending=False)

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
def cramer_coeficiente(coluna1, coluna2):
    tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
    chi2 = chi2_contingency(tabela_cruzada)[0]
    soma = np.sum(tabela_cruzada)
    valorMinino = min(tabela_cruzada.shape)-1
    cramer = np.sqrt(chi2/(soma*valorMinino))
    return cramer
```

```python
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
```
